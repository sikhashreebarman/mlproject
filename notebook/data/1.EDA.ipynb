```python
numeric_feature=[feature for feature in df.columns if df[feature].dtype !='O']  #define numerical and categorical columns
categorical_feature=[feature for feature in df.columns if df[feature].dtype=='O']
#print columns
print('we have {} numerical feature:{}'.format(len(numeric_feature),numeric_feature))
print('we have {} categorical feature:{}'.format(len(categorical_feature),categorical_feature))

```

```python
df.head(2)
```

adding columns for totral score and average

```python
df['total score']=df['math score']+df['reading score']+df['writing score']
df['average']=df['total score']/3
df.head()
```

```python
reading_full=df[df['reading score']==100]['average'].count()
print(f'number of reading full mark student: {reading_full}')
```

```python
reading_less_20=df[df['reading score']<=20]['average'].count()
print(f'num less than 20:{reading_less_20}')
```

```python
fig,axs=plt.subplots(1,2,figsize=(15,7))
plt.subplot(121)
sns.histplot(data=df,x='average',bins=30,kde=True,color='g')
plt.subplot(122)
sns.histplot(data=df,x='average',kde=True,hue='gender')
plt.show()
```
